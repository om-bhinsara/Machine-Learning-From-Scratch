## Now for smoker, time and sex we will use label encoding and for day we will use one hot encoding

```python
df.columns
```

```python
# independnent and dependent features
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train.head()
```

```python
# Feature Encoding (Label Encoding And Onehot Encoding)
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
X_train['sex']=le1.fit_transform(X_train['sex'])
X_train['smoker']=le2.fit_transform(X_train['smoker'])
X_train['time']=le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex']=le1.transform(X_test['sex'])
X_test['smoker']=le2.transform(X_test['smoker'])
X_test['time']=le3.transform(X_test['time'])
```

```python
X_test.head()
```

```python
# One-Hot Encoding using ColumnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'),[3])], remainder='passthrough')
```

```python
ct = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(drop='first', sparse_output=False), ['day'])
    ],
    remainder='passthrough'
)
```

```python
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import pandas as pd
X_train = pd.DataFrame(
    X_train,
    columns=ct.get_feature_names_out()
)

X_test = pd.DataFrame(
    X_test,
    columns=ct.get_feature_names_out()
)
```

```python
X_train.tail()
```

```python
## SVR--Support Vector Regression
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
## Hyperparameter Tuning using GridSearch CV
```

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,grid_prediction))
print(mean_absolute_error(y_test,grid_prediction))
```

```python
import matplotlib.pyplot as plt

y_pred = grid.predict(X_test)

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         linestyle='--')
plt.show()
```
